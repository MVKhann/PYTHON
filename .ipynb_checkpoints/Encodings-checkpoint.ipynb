## Dummy Variables & One-Hot Encoding

```python
import pandas as pd
```

```python
tips_df = pd.read_csv(r"tips.csv")
tips_df
```

```python
dummy_df = pd.get_dummies(tips_df)
dummy_df
```

```python
pd.get_dummies(tips_df, drop_first=True)
```

## One-Hot Encoding with Scikit-learn

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
oh_enc = OneHotEncoder(sparse=False)
```

```python
oh_enc_arr = oh_enc.fit_transform(tips_df[['gender','smoker','day','time']])
oh_enc_arr
```

```python
dummy_df.keys()
```

```python
dummy_df.keys()
oh_enc_df = pd.DataFrame(oh_enc_arr, columns=['gender_Female', 'gender_Male', 'smoker_No',
       'smoker_Yes', 'day_Fri', 'day_Sat', 'day_Sun', 'day_Thur',
       'time_Dinner', 'time_Lunch'])
oh_enc_df
```

```python
oh_enc_arr
```

## Label and Ordinal encoding

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
df=pd.read_csv('ames.csv')
```

```python
df
```

```python
df2=df[['KitchenQual','BldgType']]
df2
```

```python
le=LabelEncoder()
```

```python
le.fit_transform(df['BldgType'])
```

```python
df2['BldgType_enc']=le.fit_transform(df['BldgType'])
```

```python
df2.head(5)
```

```python
df['BldgType'].value_counts()
```

```python
df['KitchenQual'].value_counts()
```

```python
# EX==Excelent
# Gd==Good
# TA==Typical/Average
# Fa=Fair
```

```python
order={"Ex":4,"Gd":3,"TA":2,"Fa":1}
```

```python
df2["KitchenQual_en"]=df2["KitchenQual"].map(order)
```

```python
df2
```
