```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("IMDB-MOVIE-DATA.csv")
df
```

```python
year = df['Year'].unique()
size = year.size
movies= [0]*year.size

for i in range(size):
    movies[i]= df[df['Year']==year[i]].sum()

df.groupby(['Genre'])['Revenue (Millions)'].sum().plot(kind='line', figsize=(15,8), fontsize=13, color='blue')
plt.ylabel('Number of movies')
plt.xlabel('Genre')
plt.title("Total Revenue By Years")
plt.legend()
plt.show()
```
