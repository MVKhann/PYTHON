```python
import pandas as pd

df = pd.read_csv("COVID-19.csv")
df = df.dropna()
df
```

# Part 1

```python
def q1(date):    
    country = df["Country Name"].unique()
    s1 = country.size
    confirm = [0]*s1
    death = [0]*s1
    for i in range(s1):
        confirm[i] = df[(df["day"]==date)&(df["Country Name"]==country[i])]["Confirmed"].max()
        death[i] = df[(df["day"]==date)&(df["Country Name"]==country[i])]["Deaths"].max()
    print("CONFIRMED:")
    print(pd.DataFrame(country,confirm).sort_index(ascending=False).head(20))
    print("DEATHS:")
    print(pd.DataFrame(country,death).sort_index(ascending=False).head(20))
```

```python
date = input("Enter the date (y-m-d): ")
q1(date)
```

# Part 2

```python
def q2(date1,date2):
    country = df["Country Name"].unique()
    s1 = country.size
    confirm = [0]*s1
    death = [0]*s1
    for i in range(s1):
        confirm[i] = df[(df["day"]>=date1)&(df["day"]<=date2)&(df["Country Name"]==country[i])]["Confirmed"].sum().max()
        death[i] = df[(df["day"]>=date1)&(df["day"]<=date2)&(df["Country Name"]==country[i])]["Deaths"].sum().max()
    print("CONFIRMED:",date1,"-",date2)
    t = pd.DataFrame(country,confirm).sort_index(ascending=False).head(10)
    print(t)
    print("DEATHS:",date1,"-",date2)
    print(pd.DataFrame(country,death).sort_index(ascending=False).head(10))
```

```python
# Find the starting and ending days of the longest spread period for a given country. The
# spread period is defined as the period where daily new cases tend to increase. They may
# contain days where new confirmed cases were relatively lower or none at all.
# For example, [5, 2, 9, 16, 11, 27, 14, 45, 11] has a longest spread period elapsing 7 days
# from day 2(2) to day 8(45) and elements are [2,9,16,27,45] or it may be elapsing 8 days
# from day 1(5) to day (8) and elements are [5,9,16,27,45] both have same number of
# elements i.e. 5.
import numpy as np
def q3(country):
    cases = df[df['Country Name']==country]['Confirmed']
    cases = np.array(cases)
    s2 = cases.size #size of the array to be checked for longest series
    k=0 #index of the longest series array
    cntr=0 #counts the range in the longest series array
    indexes = [] #array the will store the indexes of the longest series element
    values = [] #values will store the elements of the longest array
    os_size=0 #the size of the elements in the array
    ns_size=0 #the temp size of the elements in the array
    b=False
    breakit= False
    
    for i in range(s2):
        b=False
        for j in range(s2-1):
            if(b==False):
                indexes.append(j)
                b=True
            if(cases[j]<cases[j+1]):
                indexes.append(j+1)
                cntr+=1
                ns_size+=1
            else:
                j+=1
                cntr+=1
            c=0
            a=0
            for k in range(j,s2-1):
                a+=1
                if(cases[j]>=cases[k+1]):
                    c+=1
            if(c==s2-1-j):
                breakit=True
                break
        if(breakit):
            break
        if(os_size<ns_size):
            os_size = ns_size
            ns_size = 0
        else:
            cntr=0
            indexes = []
            b=False

    print(indexes)
    s1 = len(indexes)
    values.append(cases[indexes[0]])
    k=0
    for i in range(s1-1):
         if(values[k]<cases[indexes[i+1]]):
            values.append(cases[indexes[i+1]])
            k+=1

    print("Largest spread : ",cntr)
    print("starting index and value : ",indexes[0],values[0])
    print("Ending index and value : ",indexes[len(indexes)-1],values[len(values)-1])
    print(values)
```

```python
q3("Afghanistan")
```
