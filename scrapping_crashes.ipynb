```python
from bs4 import BeautifulSoup
from urllib.request import urlopen
```

```python
page = urlopen("http://www.planecrashinfo.com/database.htm")
```

```python
html = page.read().decode("utf-8")
```

```python
soup = BeautifulSoup(html, 'html.parser')
```

```python
year_links = soup.find_all('table')[1].find_all('a')
```

```python
# year_links[0].text
# print(year_links[0]['href'])

def scrap_info(url) -> BeautifulSoup:
    print("http://www.planecrashinfo.com"+url)
    pagee = urlopen("http://www.planecrashinfo.com"+url)
    htmll = pagee.read()
    soupp = BeautifulSoup(htmll, 'html.parser')
    return soupp
```
