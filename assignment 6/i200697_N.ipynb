# Question 1

Scrape 2 pages of the reviews for a single product listing using product URL.<br /> Note: Product must have more than 1 review page. The code must be generic and should work for every product listing URL, with proper checks in place.

```python
url = 'https://www.daraz.pk/products/waterproof-mobile-cover-i124818547-s1282674849.html?spm=a2a0e.searchlist.list.27.4d6b4daaHJKmug&search=1'
```

```python
options = Options()

driver = webdriver.Chrome(ChromeDriverManager().install(), options=options)

names = []
review_text = []
driver.get(url)
page=1
for item in driver.find_elements_by_class_name('pdp-mod-review'):
    while(page<=2):
        for person in driver.find_elements_by_class_name('item'):
            try:
                name = person.find_element_by_class_name('middle').text[3:].replace('Verified Purchase','')
            except ValueError:
                name = 'anonymous'
            names.append(name)

            try:
                rev = person.find_element_by_class_name('content').text
            except ValueError:
                rev = 'NOT AVAILABLE'
            review_text.append(rev)
            
        driver.find_element_by_xpath('//*[@id="module_product_review"]/div/div[3]/div[2]/div/div/button[2]').click()

        time.sleep(5)
        page+=1
driver.close()
            
for i in range(len(names)):
    print(f'Name:, {names[i]}, Review:, {review_text[i]}')
```

```python
df = pd.DataFrame({'Name':names, 'Review':review_text})
df
```

# Question 2

```python
query = 'mobile covers'
```

```python
query = query.replace(' ','%20')
url = f'https://www.daraz.pk/catalog/?from=input&q={query}'
print(url)

options = Options()

driver = webdriver.Chrome(ChromeDriverManager().install(), options=options)

names = []
reviews = []
stars = []
prices = []
countries = []

driver.get(url)
i=0
xp=0
k=0
n=1
while(i<80):
    for product in driver.find_elements_by_class_name('c2prKC'):
        if k != xp:
            k+=1
            continue
           
        k=0
        #nameee
        try:
            name = product.find_element_by_class_name('c16H9d').text
        except ValueError:
            name = 'Not Available'
        names.append(name)

        #priceee
        try:
            price = int(product.find_element_by_class_name('c13VH6').text[3:].replace(',',''))
        except ValueError:
            price = 0
        prices.append(price)
        #reviews count
        try:
            rev = int(product.find_element_by_class_name('c3XbGJ').text[1:-1])
        except ValueError:
            rev = 0
        reviews.append(rev)
        #picking up the count
        xp+=1
        #//*[@id="root"]/div/div[2]/div[1]/div/div[1]/div[2]/div[5]/div/div
        
        
        driver.find_element_by_xpath(f'//*[@id="root"]/div/div[2]/div[1]/div/div[1]/div[2]/div[{xp}]/div/div').click()
        time.sleep(5)
        i+=1
        #for removing pop up
        try:
            pop = driver.find_element_by_xpath('/html/body/div[6]/div/div[2]/div/span').click()
            time.sleep(5)
        except:
            pop = True
            
        for p in driver.find_elements_by_class_name('pdp-mod-review'):   
            for sta in p.find_elements_by_class_name('mod-rating'):
                try:
                    rat = sta.find_element_by_class_name('score-average').text
                except ValueError:
                    rat = '0'
        stars.append(rat)
        driver.back()
        break
    if(i>=79):
        continue
    if(i>n*39):
        driver.find_element_by_xpath('//*[@id="root"]/div/div[2]/div[1]/div/div[1]/div[3]/div/ul/li[9]/a').click()
        xp=0
        n+=1
    else:
        continue
driver.close()
```

```python
df = pd.DataFrame({'Name':names, 'Prices':prices,'Stars': stars, 'Reviews' : reviews})
df
```

# Question 3

Which Products are on Flash sale? Display Product name, Actual Price, Discounted Price, Discount Percentage, Rating, No. of Reviews, Top 3 reviews displayed.

Display output must be like this. Products may vary.

```python
#Run this cell to see output
# pd.read_csv('sample.csv')
#url = f'https://www.daraz.pk/'
url = 'https://www.daraz.pk/flash-sale/'

print(url)

options = Options()

driver = webdriver.Chrome(ChromeDriverManager().install(), options=options)

names = []
prices = []
disc_prices = []
disc_percentage = []

driver.get(url)

# driver.find_element_by_xpath('//*[@id="hp-flash-sale"]/div[2]/div[1]/a').click()
# time.sleep(5)

for prod in driver.find_elements_by_class_name('item-list-content'):
    for item in driver.find_elements_by_class_name('unit-content'):
        try:
            name = item.find_element_by_class_name('sale-title').text
        except NoSuchElementException:
            name = 'Not Available'

        #priceee
        try:
            price = int(item.find_element_by_class_name('origin-price-value').text[3:].replace(',',''))
        except ValueError:
            price = 0

        #discounted price
        try:
            dprice = int(item.find_element_by_class_name('sale-price').text[3:].replace(',',''))
        except ValueError:
            dprice = 0

        try:
            dpercent = int(item.find_element_by_class_name('discount').text[1:-1])
        except NoSuchElementException:
            dpercent = 0

        prices.append(price)
        names.append(name)
        disc_prices.append(dprice)
        disc_percentage.append(dpercent)
driver.close()
df=pd.DataFrame({'Name':names,'Price':price,'Discounted Price':disc_prices,'Discount Percentage':disc_percentage})
df
```
